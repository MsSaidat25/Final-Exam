```python
import pandas as pd
import numpy as np
distance_data = {
    'point': ['b', 'c', 'd', 'e', 'f', 'g', 'h'],
    'a': [5, 8, 6, 7, 4, 4, 2],
    'b': [0, 0, 0, 0, 0, 0, 0],
    'c': [0, 6, 0, 0, 0, 0, 0],
    'd': [0, 5, 3, 0, 0, 0, 0],
    'e': [0, 5, 1, 2, 0, 0, 0],
    'f': [0, 5, 4, 1, 3, 0, 0],
    'g': [0, 1, 5, 4, 4, 4, 0],
    'h': [0, 4, 6, 4, 5, 2, 3]
}

df_dist = pd.DataFrame(distance_data)
df_dist.set_index('point', inplace=True)



```

```python
# Create a symmetric distance matrix
points = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
n = len(points)
dist_matrix = np.zeros((n, n))

# Fill the matrix
point_to_idx = {p: i for i, p in enumerate(points)}

# From the table
distances = {
    ('a', 'b'): 5, ('a', 'c'): 8, ('a', 'd'): 6, ('a', 'e'): 7,
    ('a', 'f'): 4, ('a', 'g'): 4, ('a', 'h'): 2,
    ('b', 'c'): 6, ('b', 'd'): 5, ('b', 'e'): 5, ('b', 'f'): 5,
    ('b', 'g'): 1, ('b', 'h'): 4,
    ('c', 'd'): 3, ('c', 'e'): 1, ('c', 'f'): 4, ('c', 'g'): 5,
    ('c', 'h'): 6,
    ('d', 'e'): 2, ('d', 'f'): 1, ('d', 'g'): 4, ('d', 'h'): 4,
    ('e', 'f'): 3, ('e', 'g'): 4, ('e', 'h'): 5,
    ('f', 'g'): 4, ('f', 'h'): 2,
    ('g', 'h'): 3
}
```

```python
# Fill symmetric matrix
for (p1, p2), dist in distances.items():
    i, j = point_to_idx[p1], point_to_idx[p2]
    dist_matrix[i][j] = dist
    dist_matrix[j][i] = dist

# Create DataFrame for better visualization
dist_df = pd.DataFrame(dist_matrix, index=points, columns=points)
print("\nComplete Symmetric Distance Matrix:")
print(dist_df)



```

```python

# Current cluster: (c, e)
cluster = ['c', 'e']
remaining_points = [p for p in points if p not in cluster]

print(f"\nCurrent cluster: {cluster}")
print(f"Remaining points: {remaining_points}")



```

```python

# Calculate distance from each remaining point to the cluster (c, e)
cluster_distances = {}

for point in remaining_points:
    p_idx = point_to_idx[point]

    # Distance to each point in cluster
    distances_to_cluster = []
    for cluster_point in cluster:
        c_idx = point_to_idx[cluster_point]
        dist = dist_matrix[p_idx][c_idx]
        distances_to_cluster.append(dist)

    # Single linkage = minimum distance
    min_dist = min(distances_to_cluster)
    cluster_distances[point] = min_dist

    print(f"\nPoint {point}:")
    print(f"  Distance to c: {dist_matrix[p_idx][point_to_idx['c']]}")
    print(f"  Distance to e: {dist_matrix[p_idx][point_to_idx['e']]}")
    print(f"  Single Linkage Distance: min({distances_to_cluster}) = {min_dist}")


# Sort by distance
sorted_distances = sorted(cluster_distances.items(), key=lambda x: x[1])






```

```python

# Find the point with minimum distance
next_point = sorted_distances[0][0]
min_distance = sorted_distances[0][1]

print(f" {next_point}")





```

```python
# DBSCAN Parameters
epsilon = 2
minPts = 2

# Find epsilon-neighborhood for each point
neighborhoods = {}
for point in points:
    p_idx = point_to_idx[point]
    neighbors = [point]
    for other_point in points:
        if other_point != point:
            o_idx = point_to_idx[other_point]
            if dist_matrix[p_idx][o_idx] <= epsilon:
                neighbors.append(other_point)
    neighborhoods[point] = neighbors


# Classify points as Core, Border, or Noise

core_points = []
border_points = []
noise_points = []

for point in points:
    count = len(neighborhoods[point])
    if count >= minPts:
        core_points.append(point)
        status = "CORE"
    else:

        is_border = False
        for cp in points:
            if len(neighborhoods[cp]) >= minPts and point in neighborhoods[cp]:
                is_border = True
                break
        if is_border:
            status = "BORDER (will be determined after identifying all core points)"
        else:
            status = "NOISE (tentative)"
    print(f"\n{point}: {len(neighborhoods[point])} points in neighborhood \u2192 {status}")

# Final classification after identifying all core points

final_core_points = []
final_border_points = []
final_noise_points = []

for point in points:
    if len(neighborhoods[point]) >= minPts:
        final_core_points.append(point)

# Need to re-iterate for border/noise after core points are definitively identified
for point in points:
    if point not in final_core_points:
        is_border = False
        for cp in final_core_points:
            if point in neighborhoods[cp]:
                is_border = True
                break
        if is_border:
            final_border_points.append(point)
        else:
            final_noise_points.append(point)

print(f"\nCore Points: {final_core_points}")
print(f"Border Points: {final_border_points}")
print(f"Noise Points: {final_noise_points}")

# Step 3: Check Density Reachability

def is_directly_density_reachable(p, q, core_points_list, neighborhoods_dict):
    """
    p is directly density-reachable from q if:
    1. q is a core point
    2. p is in N_ε(q)
    """
    return q in core_points_list and p in neighborhoods_dict[q]

def is_density_reachable(p, q, core_points_list, neighborhoods_dict):
    """
    p is density-reachable from q if there exists a chain of points
    p1, p2, ..., pn where p1=q, pn=p, and each pi+1 is directly
    density-reachable from pi
    """
    if p == q:
        return True

    visited = set()
    queue = [q]
    visited.add(q)

    while queue:
        current = queue.pop(0)

        # If current is not a core point, it cannot expand the chain
        if current not in core_points_list:
            continue

        for neighbor in neighborhoods_dict[current]:
            if neighbor not in visited:
                if neighbor == p:
                    return True
                visited.add(neighbor)
                queue.append(neighbor)

    return False
```

```python
data = pd.DataFrame({
    "gender": ["male","male","female","female","male","female","female","male","female"],
    "age": ["young","young","young","teenager","young","young","elder","middle age","elder"],
    "income": ["medium","low","low","medium","high","medium","high","medium","medium"],
    "play": ["yes","no","no","no","yes","no","yes","yes","yes"],
    "count": [30,20,30,20,15,20,13,10,4]
})

# Expand the data so that counts become actual instances
df = data.loc[data.index.repeat(data["count"])].reset_index(drop=True)

# The query tuple
q_gender = "female"
q_age = "teenager"
q_income = "low"

# Compute P(no)
p_no = (df["play"] == "no").mean()

# Compute needed conditional probabilities
p_gender_female_no = ((df["gender"] == q_gender) & (df["play"] == "no")).sum() / (df["play"] == "no").sum()
p_age_teen_no = ((df["age"] == q_age) & (df["play"] == "no")).sum() / (df["play"] == "no").sum()

p_income_low_no = ((df["income"] == q_income) & (df["play"] == "no")).sum() / (df["play"] == "no").sum()


print("P(no) =", p_no)
print("P(gender = female | no) =", p_gender_female_no)
print("P(age = teenager | no) =", p_age_teen_no)
print("P(income = low | no) =", p_income_low_no)


```

```python
data = pd.DataFrame({
    "gender": ["male","male","female","female","male","female","female","male","female"],
    "age": ["young","young","young","teenager","young","young","elder","middle age","elder"],
    "income": ["medium","low","low","medium","high","medium","high","medium","medium"],
    "play": ["yes","no","no","no","yes","no","yes","yes","yes"],
    "count": [30,20,30,20,15,20,13,10,4]
})

# Query instance
q_gender = "male"
q_age = "young"
q_income = "medium"

# Find the matching row in the dataset
match = data[
    (data["gender"] == q_gender) &
    (data["age"] == q_age) &
    (data["income"] == q_income)
]

true_class = match["play"].iloc[0]
predicted_class = "no"  # given in the question

# Determine the confusion matrix category
if predicted_class == "yes" and true_class == "yes":
    result = "True Positive"
elif predicted_class == "no" and true_class == "no":
    result = "True Negative"
elif predicted_class == "yes" and true_class == "no":
    result = "False Positive"
elif predicted_class == "no" and true_class == "yes":
    result = "False Negative"
else:
    result = "Unknown"

print("True class:", true_class)
print("Predicted class:", predicted_class)
print("Outcome:", result)
```
